```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


#Leer archivo
datos = pd.read_csv("wine_cluster.csv")
print(datos)
#se elimina filas que tenga valores nulos
data = datos.dropna()

#reiniciar indice
datos = datos.reset_index(drop = True)
print(datos.head())

#caracteristicas basicas estadisticas
print(datos.describe())

#Separa las caracteristicas relevantes
x = datos[{'Alcohol','Color_Intensity'}]

#actualizar los datos
x = (x-x.mean())/x.std()

#grafico de dispercion para visualizar datos de entrada
plt.scatter(x.iloc[:,0],x.iloc[:,1])
plt.xlabel('Alcohol')
plt.ylabel('Color_Intesnity')
plt.show()

#calcular la inercia para diferentes valores de cluster
k_valores = range(2,10) #numero de cluster
inercias =[]
for k in k_valores:
  modelo =KMeans(k) #creacion del modelo
  modelo.fit(x) #prueba el modelo
  inercias.append(modelo.inertia_)

#Graficar la curva de codo
plt.scatter(k_valores,inercias)
plt.xlabel('Numero de cluster')
plt.ylabel('Inercia')
plt.show()

#calcular la inercia para diferentes valores de cluster con silueta
k_valores = range(2,10) #numero de cluster
inercias =[]
siluetas=[]
for k in k_valores:
  modelo =KMeans(k) #creacion del modelo
  modelo.fit(x) #prueba el modelo
  inercias.append(modelo.inertia_)
  silueta=silhouette_score(x,modelo.labels_)#calcula coeficiente de silueta cuando hay uno, dos o tres
  siluetas.append(silueta)

#Graficar los valores de silueta
plt.scatter(k_valores,siluetas)
plt.xlabel('Numero de cluster')
plt.ylabel('Valor coeficiente de silueta')
plt.show()
print(siluetas[2])

#crear modelo y entrenar
modelo =KMeans(5)
modelo.fit(x) #calcular los centroides del modelo
centroide = modelo.cluster_centers_
etiquetas = modelo.labels_
print(centroide)
print(etiquetas)


#Grafico de los cluster
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=etiquetas,cmap ='viridis')
plt.scatter(centroide[:,0],centroide[:,1],marker ='x', color ='red')
plt.xlabel('Alcohol')
plt.ylabel('Color_Intesnity')
plt.show()

#evaluacion de modelo con las metricas/ este modelo fue entrenado con datos normalizados por lo tanto sus dats
#calcular los coeficientes necesito los datos y coeficientes
silueta=silhouette_score(x,etiquetas) #calcula coeficiente de silueta cuando hay uno, dos o tres
Calinski=calinski_harabasz_score(x,etiquetas)
davies =davies_bouldin_score(x,etiquetas)
print(silueta)
print(Calinski)
print(davies)

#prediccion
datos_nuevos=np.array([[14.5,8]])
#normalizar
datos_nuevos = (datos_nuevos - datos_nuevos.mean())/datos_nuevos.std()
prediccion = modelo.predict(datos_nuevos)
print("El clouster es: ", prediccion)

```
